```python
from collections import defaultdict
def mapper():
    test1=open("train.tsv")
    tscore=defaultdict()
    for t in test1:
        
        phraseid,sentenceid,phrase,sentiment=t.split('\t')
        tscore[phrase]=(sentiment[:-1])
    return tscore
def test(tscore):
    test=open("test.tsv")
    columns=defaultdict()
    for t in test:
        phraseid,sentenceid,phrase=t.split('\t')
#         print phrase
        try:
            columns[phraseid]=tscore[phrase[:-1]]
        except:
            columns[phraseid]=2
    print columns
    
    
tscore=mapper()
test(tscore)
```

```python
def parseafin():
    afinnfile=open("AFINN-111.txt")
    scores={} # initialize an empty dictionary
    for line in afinnfile:
        #note that when you have a multiple variable set. it will iterate over the variable for each line
        term, score = line.split("\t") # The file is tab-delimted. "\t" means "tab character".
        scores[term]=int(score) #convert the score to an integer.
    print scores.items() # print every (term,score) pair in the dictionary
    return scores
scores=parseafin()
```
